### Scatter Plot
1- A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables.

2- The position of each dot on the horizontal and vertical axis indicates values for an individual data point.

3- Scatter plots are used to observe relationships between variables.

```python
plt.scatter(df.crime_rate,df.percent_senior)
plt.show()
```

```python
plt.scatter(df.crime_rate,df.percent_city)
plt.show()
```
